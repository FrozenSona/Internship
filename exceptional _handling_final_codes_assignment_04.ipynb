1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# intantiating webdriver
driver=webdriver.Chrome("chromedriver.exe")
# Loading url to the chrome driver
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
driver.maximize_window()
```

```python
#creating list for rank, uploader)name, video name, no of views, date of upload
rank = []
video_names = []
artist = []
views = []
upload_dates = []
```

```python
# getting ranks of the videos
try:
    rank_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')[:30]
    rank_obj
    for i in rank_obj:
        rank.append(i.text.replace('.',' '))
except NoSuchElementException:
        rank.append('-')
print(rank)
print("=======================================")
# getting names of the videos
try:
    videoname_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')[:30]
    videoname_obj
    for i in videoname_obj:
        video_names.append(i.text)
except NoSuchElementException:
        video_names.append('-')
print(video_names)
print("============================================")
#getting names of artist
try:
    uploader_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')[:30]
    for i in uploader_obj:
        artist.append(i.text)
except NoSuchElementException :
        artist.append('-')
print(artist)
print("==============================================")
#getting numbers of views of the videos
try:
    view_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')[:30]
    view_obj
    for i in view_obj:
        views.append(i.text)
except NoSuchElementException :
        views.append('-')
print(views)    
print("==============================================================")    
#getting upload date
try:
    date_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')[:30]
    date_obj
    for i in date_obj:
        upload_dates.append(i.text)
except NoSuchElementException :
        upload_dates.append('-')
print(upload_dates)
print("==============================================")
```

```python
print(len(views),len(rank),len(video_names),len(artist),len(upload_dates))
```

```python
# as we see length are same ,we can create dataframe "dataframe created"
WIKI_VIEWS=pd.DataFrame({})
WIKI_VIEWS['RANK']=rank
WIKI_VIEWS['VIDEOS_NAME']=video_names
WIKI_VIEWS['ARTIST']=artist
WIKI_VIEWS['VIEWS']=views
WIKI_VIEWS['UPLOAD_DATE']=upload_dates
WIKI_VIEWS.reset_index(drop=True,inplace=True)
WIKI_VIEWS.index+=1
WIKI_VIEWS
```

```python
url='https://github.com/'
driver=webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.maximize_window()
```

```python
time.sleep(3)
three=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[1]/div[2]/button').click()
time.sleep(3)
explorer=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary').click()
time.sleep(3)
trending=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a').click()
time.sleep(3)
```

```python
#Creating empty list
repo_urls = []
rep_title = []
Description = []
Contributors = []
Language_used = []
l = []
```

```python
#Fetching urls for each repository
repository = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']//a")
for i in repository:
    repo_urls.append(i.get_attribute("href"))
    
#Scraping data of repository title
title = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
for i in title:
    rep_title.append(i.text)
rep_title
```

```python
#Scraping data from every repository page
for i in repo_urls:
    driver.get(i)
    time.sleep(3)
    
    
        #Scraping data of decription
    try:
        desc = driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[1]/div/p")
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
        
        #scraping contributors count
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors.append(count.text)
    except NoSuchElementException:#handling no such element exception
        Contributors.append('No details available')
    time.sleep(2)
    
        #fetching Languages used
     #scraping languages used   
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:    
        for i in languages:
            l.append(i.text)
    else:
        l.append('No languages used')
    Language_used.append(l)
```

```python
print(len(repo_urls),len(rep_title),len(Description),len(Contributors),len(Language_used),len(l))
```

```python
GITHUB=pd.DataFrame({})
GITHUB['TITLE']=rep_title
GITHUB['DESCRIPTION']=Description
GITHUB['CONTRIBUTION']=Contributors
GITHUB['LANGUAGE']=Language_used
GITHUB.reset_index(drop=True,inplace=True)
GITHUB.index+=1
GITHUB
```
